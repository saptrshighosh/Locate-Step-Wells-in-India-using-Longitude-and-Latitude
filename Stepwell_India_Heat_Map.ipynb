```python
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point

# Create a GeoDataFrame for the stepwell locations
data = {
    'name': [
        "Chand Baori",
        "Rani-ki-Vav",
        "Agrasen Ki Baoli",
        "Adalaj Vav",
        "Rajon Ki Baoli",
        "Surya Kund Stepwell"
    ],
    'latitude': [
        27.007222,  # Chand Baori
        23.858889,  # Rani-ki-Vav
        28.626111,  # Agrasen Ki Baoli
        23.16691,   # Adalaj Vav
        28.520261,  # Rajon Ki Baoli
        23.583806   # Surya Kund Stepwell
    ],
    'longitude': [
        76.606389,  # Chand Baori
        72.101667,  # Rani-ki-Vav
        77.225000,  # Agrasen Ki Baoli
        72.58024,   # Adalaj Vav
        77.183411,  # Rajon Ki Baoli
        72.132686   # Surya Kund Stepwell
    ]
}

gdf = gpd.GeoDataFrame(data, geometry=gpd.points_from_xy(data['longitude'], data['latitude']))

# Create a base map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
india = world[world.name == 'India']

# Plot the map
fig, ax = plt.subplots(figsize=(10, 10))
india.plot(ax=ax, color='lightgrey')

# Plot the stepwells
gdf.plot(ax=ax, color='blue', markersize=50, label='Stepwell')

# Annotate the stepwells
for x, y, label in zip(gdf.geometry.x, gdf.geometry.y, gdf['name']):
    ax.text(x, y, label, fontsize=8, ha='right')

plt.title('Stepwell Locations in India')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.show()
```
